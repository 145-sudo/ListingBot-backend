```python
from woocommerce import API
import pandas as pd
import numpy as np
import os
```

```python
# Initialize the WooCommerce API client
wcapi = API(
    url="https://stltacticals.com",
    consumer_key=os.getenv("WP_CONSUMER_KEY"),
    consumer_secret=os.getenv("WP_CONSUMER_SECRET"),
    version="wc/v3"
)
```

```python

# Fetch products (up to 100 per page)
response = wcapi.get("products", params={"per_page": 10})

# Convert response to JSON (Python dictionary)
products = response.json()
```

```python
# Get total products count from response headers
total_products = int(response.headers.get('X-WP-Total', 0))

print(f"Total products count: {total_products}")
```

```python

# Print product details
for product in products:
    print(f"ID: {product['id']}, Name: {product['name']}, Price: {product['price']}")
```

```python
def get_products(wcapi, page=1, per_page=100):
    products = []
    try:
        response = wcapi.get("products", params={"page": page, "per_page": per_page})
        products = response.json()
        return products
    except Exception as e:
        print(f"Error fetching products on Page:{page}, Per_Page:{per_page}")
        print(e)
        return []

def get_all_products(wcapi, total_products):
    products = []
    page = 1
    per_page = 100
    while True:
        products.extend(get_products(wcapi, page, per_page))
        print(f"Fetched {len(products)} products, Page:{page}, Per_Page:{per_page}")
        if len(products) >= total_products:
            break
        page += 1
    return products
```

```python
products = get_all_products(wcapi, total_products)
```

```python
df = pd.DataFrame(products)
```
