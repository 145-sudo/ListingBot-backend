```python
from util.gsheet import get_sheet_data

sheet_id = "1ZE7BeE26iM43x_ympaIgOdl6lXWrwXg-PZZAKuIBEIw"
sheet_name = "Sheet20"

ssis = get_sheet_data(sheet_id, sheet_name, df=True)
```

```python
ssis
```

```python

# Load store products
store_products = products  # your DataFrame from earlier

# Load SSI products
ssi_products = ssis

# Normalize names for comparison
def normalize_name(name):
    return str(name).strip().lower()
```

```python

store_products['norm_name'] = store_products['name'].apply(normalize_name)
ssi_products['norm_name'] = ssi_products['Name'].apply(normalize_name)

# Merge on normalized name
comparison = pd.merge(
    ssi_products, store_products,
    on='norm_name',
    how='left',
    suffixes=('_ssi', '_store')
)
```

```python
# Find matches and non-matches
matched = comparison[~comparison['id'].isna()]
unmatched = comparison[comparison['id'].isna()]

print("Matched products:", len(matched))
print("Unmatched SSI products:", len(unmatched))
```
